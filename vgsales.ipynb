```python

# DATA UNDERSTANDING

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('vgsales.csv')

# Identifikasi Variabel
print("Variabel-variabel dalam dataset:")
print(df.columns)
```

```python
# Distribusi Data
print("\nDistribusi data penjualan video game:")
print(df['NA_Sales'].describe())

# Analisis distribusi data penjualan video game
plt.figure(figsize=(10,6))
plt.hist(df['NA_Sales'], bins=50)
plt.title('Distribusi Penjualan Video Game di Amerika Utara')
plt.xlabel('Penjualan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Kualitas Data
print("\nKualitas data:")
print(df.isnull().sum())
```

```python
# Missing Value
print("\nMissing value:")
print(df.isnull().sum())
```

```python
# Outlier
print("\nOutlier:")
plt.figure(figsize=(10,6))
plt.boxplot(df['NA_Sales'])
plt.title('Boxplot Penjualan Video Game di Amerika Utara')
plt.show()
```

```python
# Korelasi
print("\nKorelasi antara variabel-variabel:")
print(df.select_dtypes(include=['int64', 'float64']).corr())
```

```python
# Struktur Data
print("\nStruktur data:")
print(df.head())
```

```python

# DATA PREPARATION

import pandas as pd

# Load dataset
df = pd.read_csv('vgsales.csv')

# Cek jumlah missing value sebelum mengatasi missing value
print("Jumlah missing value sebelum mengatasi missing value:")
print(df.isnull().sum())

# Mengatasi missing value
df.dropna(inplace=True)

# Cek jumlah missing value setelah mengatasi missing value
print("\nJumlah missing value setelah mengatasi missing value:")
print(df.isnull().sum())

# Cek dataframe setelah mengatasi missing value
print("\nDataframe setelah mengatasi missing value:")
print(df.head())

# Cek statistik deskriptif
print("\nStatistik deskriptif:")
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('vgsales.csv')

# Cek statistik deskriptif sebelum mengatasi outlier
print("Statistik deskriptif sebelum mengatasi outlier:")
print(df['NA_Sales'].describe())

# Mengatasi outlier
df = df[df['NA_Sales'] < 10]

# Cek statistik deskriptif setelah mengatasi outlier
print("\nStatistik deskriptif setelah mengatasi outlier:")
print(df['NA_Sales'].describe())

# Cek distribusi data setelah mengatasi outlier
plt.hist(df['NA_Sales'], bins=50)
plt.title('Distribusi Setelah Mengatasi Outlier')
plt.show()
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('vgsales.csv')

# Cek tipe data sebelum mengubah
print("Tipe data sebelum mengubah:")
print(df['Year'].dtype)

# Mengubah tipe data
df['Year'] = pd.to_numeric(df['Year'])

# Cek tipe data setelah mengubah
print("\nTipe data setelah mengubah:")
print(df['Year'].dtype)

# Cek statistik deskriptif
print("\nStatistik deskriptif:")
print(df['Year'].describe())
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('vgsales.csv')

# Cek jumlah baris sebelum menghapus duplikat
print("Jumlah baris sebelum menghapus duplikat:", len(df))

# Menghapus duplikat
df.drop_duplicates(inplace=True)

# Cek jumlah baris setelah menghapus duplikat
print("Jumlah baris setelah menghapus duplikat:", len(df))
```

```python
import matplotlib.pyplot as plt

# Plot histogram
plt.hist(df['NA_Sales'], bins=50)
plt.title('Distribusi Penjualan Video Game di Amerika Utara')
plt.xlabel('Penjualan')
plt.ylabel('Frekuensi')
plt.show()

# Plot scatter plot
plt.scatter(df['Year'], df['NA_Sales'])
plt.title('Hubungan antara Tahun dan Penjualan Video Game di Amerika Utara')
plt.xlabel('Tahun')
plt.ylabel('Penjualan')
plt.show()
```
